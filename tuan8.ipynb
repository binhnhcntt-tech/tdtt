### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def dx(lst):
    if lst:
        for i, x in enumerate(lst):
            if x==0:
                lst.pop(i)
                lst.insert(0, 0)
        return lst
    return None
assert dx([1,2,3,0,0,4]) ==[0,0,1,2,3,4]
print("pass all")
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

```python
def gn(lst):
    if lst:
        a= []
        for i in lst:
            if i not in a:
                a.append(i)
        return a
    return None
assert gn([1,2,2]) ==[1,2]
print("pass ")
```
