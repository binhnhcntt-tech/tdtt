Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 


```python
a, b = map(int, input().split())
for i in range(1,min(a,b)+1):
    if (a%i)==(b%i):
        print(i)

```

Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25

```python
a=int(input())
i=0
while i*i<=a:
    if i*i==a:
        print("so chinh phuong")
        break
    i+=1
else:
    print("ko")
```

Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 


```python
a = int(input())
T=0
for i in range(1,a):
    if a%i==0:
        T+=i
if T== a:
    print("so hoan hao")
else:
    print("ko phai")
```

Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
S = 0
k = 0
while S +(k+1) < n:
    k += 1
    S += k
print("k =", k)
```

Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 


```python
a = int(input())
s=1
i=1
while s<=a:
    i+=1
    s+=1/i
print(i)

```
