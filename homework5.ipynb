Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def stp(x)->bool:
    for i in range(1,x):
        if i*i==x:
            return True
    else:
        return False
x=int(input())
if stp(x):
    print(True)
else:
    print(False)
```

 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def sds(x)->bool:
    while x<0:
        x=-x
    t=0
    v=x
    while x>0:
        c=x%10
        t=t*10+c
        x//=10
    if v==t:
        return True
n=int(input())
if sds(n):
    print(True)
else:
    print(False)

```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def dnt(s):
    if "." in s:
        return "so thuc"
    else:
        return "so nguyen"
n=input()
if dnt(n)=="so nguyen":
    print("int")
elif dnt(n)=="so thuc":
    print("float")
else:
    print("neither")


```

Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def iopoo(a,b,c):
    if a==b*c or b==a*c or c==a*b:
        return True
    else:
        return False
a, b, c=map(int, input().split())
if iopoo(a,b,c):
    print(True)
else:
    print(False)
```

Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không

```python
def collinear(x1, y1, x2, y2, x3, y3):
    return (y2 - y1)*(x3 - x1) == (y3 - y1)*(x2 - x1)
```

Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def kk(list):
    cnt=0
    for i in list:
        for x in str(abs(i)):
            if x != "0":
                cnt+=1
    return cnt
list=(20, 5, 90)
print(kk(list))



```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min]

```python
def mm(a,b,c):
    if a>b and a>c:
        if b>c:
            return [a,c]
        else:
            return [a,b]
    elif b>c and b>a:
        if a>c:
            return [b,c]
        else:
            return [b,a]
    else:
        if a>b:
            return [c,b]
        else:
            return [c,a]    
print(mm(5,3,9))
```
