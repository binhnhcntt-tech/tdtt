 Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình

```python
a, b, c = map(int, input().split())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    P = a + b + c
    if a == b == c:
        loai = "tam giac deu"
    elif a == b or b == c or a == c:
        loai = "tam giac can"
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        loai = "tam giac vuong"
    else:
        loai = "tam giac thuong"
    print("day la", loai)
    print("dien tich =", S)
    print("chu vi =", P)
else:
    print("day ko phai tam giac")
```

bai 8 Giải hệ phương trình tuyến tính:
ax  by  m
cx  dy  n

```python
a, b ,c ,d, m ,n= map(float, input().split())
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c

if D != 0:
    x = Dx / D
    y = Dy / D
    print("Nghiem duy nhat:")
    print("x =", x)
    print("y =", y)
else:
    if Dx == 0 and Dy == 0:
        print("Vo so nghiem")
    else:
        print("Vo nghiem")
```

bai 9 Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây

```python
x = int(input())
h = x//3600
a = x%3600
if a != 0:
    b = a//60
    c = a%60
    print(h,"giờ",b,"phút",c, "giây")
else:
    print(h,"giờ")
```
