"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."


```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
a, b = map(int, input().split())
T = 0
for i in range(a,b+1):
    if is_prime(i):
          T += i
print(T)
```

W4A8
ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị pal

```python
n = int(input())
def reverse(n):
    return int(str(n)[::-1])
def is_palindrome(n):
    return str(n) == str(n)[::-1]
buoc = 0
while not is_palindrome(n):
    n += reverse(n)
    buoc += 1

print(buoc, n)
```

W4A9
[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number)

```python
def has_distinct_digits(n):
    s = str(n)
    return len(s) == len(set(s))
n = int(input())
found = False
i = 1
while i * i <= n:
    x = i * i
    if has_distinct_digits(x):
        print(x, end=" ")
        found = True
    i += 1
if not found:
    print("(no number)")
```
