W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
n=int(input())
def check_prime_num(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return 'khong la so nguyen to'
        else:
            return 'la so nguyen to'
print(f'{n} {check_prime_num(n)}')
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
n=int(input())

def check_perfect_num(n):
    temp=0
    for i in range(1,n):
        if n%i==0:
            temp=temp+i
    if temp==n:
            return 'la so hoan hao'
    else:
            return 'khong la so hoan hao'
print(check_perfect_num(n))
```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

Phép trừ \( - \)

W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn

```python
a=input()
b=input()

def hamming_dis(a,b):
    kcach=0
    if len(a)!=len(b):
        return 'khong the so sanh'
    for i in range(len(a)):
        if a[i]!=b[i]:
            kcach=kcach+1
    return kcach
print(hamming_dis(a,b))
```

W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
n=int(input())
def total(n):
    tong=0
    while n>0:
        chu_so=n%10
        tong=tong+chu_so
        n=n//10
    return tong
print(total(n))

```

W5A10
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 


```python
#so sanh vi tri xuat hien dau tien
s=input()
t=input()
def dang_cau_check(s,t):
    if len(s)!=len(t):
        return False
    for i in range(len(s)):
        if s.find(s[i])!=t.find(t[i]):
            return False
    return True
print(dang_cau_check(s,t))
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
a=float(input())
b=float(input())
op=input()
def compute(a,b,op):
    if op=='+':
        return a + b
    if op=='-':
        return a-b
    if op=='*':
        return a*b
    if op=='/':
        if b==0:
            return 'khong co phep chia cho 0'
        return a/b
print(compute(a,b,op))
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
a=float(input())
b=float(input())
c=float(input())
def is_right_triangle(a,b,c):
    if a<=0 or b<=0 or c<=0:
        return False
    if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        return True
    return False
print(is_right_triangle(a,b,c))
```
